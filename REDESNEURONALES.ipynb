```python
# Importamos la librería para funciones matemáticas
import numpy as np

# Definimos la función sigmoide
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Valores de entrada
x = 6          # horas que una persona pasa en el sitio
w = 1.2        # peso aprendido
b = -4         # sesgo

# Calculamos z y la probabilidad
z = w * x + b
prob = sigmoid(z)

# Mostramos los resultados
print(f"z = {z}")
print(f"Probabilidad de que compre: {prob:.4f}")

# Tomamos decisión
decision = 1 if prob >= 0.5 else 0
print(f"Decisión final: {decision} (1 = compra, 0 = no compra)")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función de error
def error(w):
    return (w - 3) ** 2

# Derivada de la función de error
def gradiente(w):
    return 2 * (w - 3)

# Parámetros
w_actual = 8              # Valor inicial del peso (lejos del mínimo)
tasa_aprendizaje = 0.1    # Cuánto avanza en cada paso
historial_w = [w_actual]  # Guardamos los valores de w

# Entrenamiento: aplicar descenso del gradiente
for i in range(20):  # 20 pasos
    grad = gradiente(w_actual)
    w_actual = w_actual - tasa_aprendizaje * grad
    historial_w.append(w_actual)

# Visualización
w_vals = np.linspace(0, 6, 100)
e_vals = error(w_vals)

plt.figure(figsize=(8, 5))
plt.plot(w_vals, e_vals, label='Función de Error E(w)')
plt.scatter(historial_w, error(np.array(historial_w)), color='red', label='Pasos del gradiente')
plt.xlabel('w')
plt.ylabel('Error')
plt.title('Descenso del Gradiente')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
# Paso 1: Creamos los datos de entrada y salida reales
x = [1, 2, 3, 4, 5]       # Entradas
y = [3, 5, 7, 9, 11]      # Salidas reales (porque y = 2x + 1)

# Paso 2: Iniciar w (peso) y b (sesgo) con cualquier valor
w = 0.0
b = 0.0

# Paso 3: Tasa de aprendizaje
lr = 0.01   # Qué tan grande es cada paso del aprendizaje
```

```python
# Paso 1: Datos reales (entrada x, salida y)
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 9, 11]  # Porque y = 2x + 1

# Paso 2: Inicializar w y b
w = 0.0
b = 0.0

# Paso 3: Tasa de aprendizaje
lr = 0.01

# Entrenamiento
for epoch in range(100):
    # Paso 4: Predicción de la neurona
    y_pred = [w * xi + b for xi in x]

    # Paso 5: Error cuadrático medio
    mse = sum((y_pred[i] - y[i])**2 for i in range(len(x))) / len(x)

    # Paso 6: Gradientes
    dw = sum(2 * x[i] * (y_pred[i] - y[i]) for i in range(len(x))) / len(x)
    db = sum(2 * (y_pred[i] - y[i]) for i in range(len(x))) / len(x)

    # Paso 7: Actualización
    w = w - lr * dw
    b = b - lr * db

    # Mostrar progreso cada 10 épocas
    if epoch % 10 == 0:
        print(f"Epoch {epoch} - Error: {mse:.4f}, w: {w:.4f}, b: {b:.4f}")

# Resultado final
print("\n🧠 Resultado final del aprendizaje:")
print(f"Peso aprendido (w): {w:.4f}")
print(f"Sesgo aprendido (b): {b:.4f}")
```

Usa una lista de datos de entrada x = [1, 2, 3, 4].

Calcula los valores reales de y usando la fórmula y = 3x + 1.

Inicializa w = 0.0 y b = 0.0.

Usa una tasa de aprendizaje lr = 0.01.

Entrena por 100 épocas con la fórmula de predicción: y_hat = w * x + b.

Calcula el error y actualiza w y b usando descenso del gradiente.

Imprime el resultado cada 20 épocas.

```python
# Neurona simple para aprender y = 3x + 1

# Datos de entrenamiento
x = [1, 2, 3, 4]
y = [4, 7, 10, 13]  # y = 3x + 1

# Inicializamos los parámetros
w = 0.0  # Peso
b = 0.0  # Sesgo
lr = 0.01  # Tasa de aprendizaje

# Entrenamiento
for epoch in range(100):
    dw = 0
    db = 0
    loss = 0

    for i in range(len(x)):
        y_hat = w * x[i] + b  # Predicción
        error = y_hat - y[i]  # Error

        dw += error * x[i]   # Gradiente para w
        db += error          # Gradiente para b
        loss += error ** 2   # Error cuadrático

    # Promediamos los gradientes
    dw /= len(x)
    db /= len(x)
    loss /= len(x)

    # Actualizamos los parámetros
    w -= lr * dw
    b -= lr * db

    # Imprimimos el progreso cada 10 épocas
    if epoch % 10 == 0:
        print(f"Epoch {epoch} - Loss: {loss:.4f}, w: {w:.4f}, b: {b:.4f}")

# Resultados finales
print("\nEntrenamiento terminado:")
print(f"Peso final (w): {w}")
print(f"Sesgo final (b): {b}")

```
